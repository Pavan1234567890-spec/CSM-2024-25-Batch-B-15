```python
df['age']=df['age'].fillna(df['age'].median())
```

```python
# Display unique values and their counts for the 'class' column
print("Unique values in 'class' column:")
print(df['class'].value_counts())
print("\nData type of 'class' column:", df['class'].dtype)
```

```python
df['sex']=df['sex'].map({
    'male':0,
    'female':1
})
```

```python
df.head()
```

```python
# Apply One-Hot Encoding to 'class', 'embarked', and 'who' columns
```

As you can see, the 'class', 'embarked', and 'who' columns have been replaced by new binary columns, which are suitable for a logistic regression model. I used `drop_first=True` to avoid multicollinearity (the 'dummy variable trap') by dropping the first category of each one-hot encoded variable.

```python
df['adult_male'].value_counts()
```

```python
df['adult_male'] = df['adult_male'].map({
    True: 1,
    False: 0
})
```

```python
df.head()
```

```python
df['embark_town'].value_counts()
```

```python
df['embarked'].value_counts()
df['class'].value_counts()
df['who'].value_counts()
```

```python
df['class'].value_counts()
```

```python
df['embarked'].value_counts()
```

```python
df=pd.get_dummies(df,columns=['class','embarked','embark_town','who'],drop_first=True)
```

```python
df
```

Now, the `alive` and `alone` columns have been converted to 0s and 1s, making them ready for your logistic regression model.

```python
# Convert 'alive' column from 'yes'/'no' to 1/0
df['alive'] = df['alive'].map({'yes': 1, 'no': 0})

# Get all boolean columns and convert them to integer (0 or 1)
boolean_columns = df.select_dtypes(include='bool').columns
for col in boolean_columns:
    df[col] = df[col].astype(int)

display(df.head())
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
df.head()
```

```python
df['pclass'].value_counts()
```

```python
numerical_cols=['age','fare','pclass']

scaler=StandardScaler()
df[numerical_cols]=scaler.fit_transform(df[numerical_cols])
```

```python
df.head()
```

```python
numerical_cols = ['age', 'fare', 'pclass', 'survived']

corr_matrix = df[numerical_cols].corr(method='pearson')

print(corr_matrix['survived'])
```

```python
df.columns
```

```python
from sklearn.feature_selection import chi2

categorical_cols = [
    'sex', 'sibsp', 'parch',
    'adult_male', 'alive', 'alone',
    'class_Second', 'class_Third',
    'embarked_Q', 'embarked_S',
    'embark_town_Queenstown',
    'embark_town_Southampton',
    'who_man', 'who_woman'
]

df_x = df[categorical_cols]
df_y = df['survived']

chi2_scores, p_values = chi2(df_x, df_y)

print(chi2_scores)
print(p_values)
```

```python
df = df.drop(columns=[
    'sibsp',
    'embarked_Q',
    'embark_town_Queenstown',
    'alive'
])
```

```python
x=df.drop('survived',axis=1)
y=df['survived']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()

model.fit(x_train, y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report

accuracy_score=accuracy_score(y_test,y_pred)
```

```python
accuracy_score
```

```python
cm = confusion_matrix(y_test, y_pred)

print("Confusion Matrix:")
print(cm)

# Classification Report
cr = classification_report(y_test, y_pred)

print("\nClassification Report:")
print(cr)
```

```python
from sklearn .model_selection import cross_val_score
scores=cross_val_score(
    model,
    x,
    y,
    cv=5,
    scoring='accuracy'
)
print(scores)
```
